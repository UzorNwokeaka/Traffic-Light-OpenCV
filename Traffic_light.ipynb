```python
# A Simple Python code that plots a traffic light using openCV lib. 

# import library
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Create the canvass 
canvass= np.zeros((1200,1200,3))

# Create the rectangular frame to house the lights
cv2.rectangle(canvass, pt1=(400,100), pt2=(800,800), color=(60,150,255), thickness=8)
cv2.rectangle(canvass, pt1=(420,120), pt2=(780,780), color=(60,150,255), thickness=5)

# Create the base of the traffic light 
cv2.rectangle(canvass, pt1=(550,800), pt2=(650,1200), color=(0,45,45), thickness=-1)

# Create the circular frames for the lights 
cv2.circle(canvass, center=(600,250), radius=85, color=(60,150, 255), thickness=5)
cv2.circle(canvass, center=(600,250), radius=80, color=(255,0, 0), thickness=-1)
cv2.circle(canvass, center=(600,450), radius=85, color=(60,150,255), thickness=5)
cv2.circle(canvass, center=(600,450), radius=80, color=(255,195,0), thickness=-1)
cv2.circle(canvass, center=(600,650), radius=85, color=(60,150, 255), thickness=5)
cv2.circle(canvass, center=(600,650), radius=80, color=(0,255,0), thickness=-1)

# Plot the canvass 
plt.imshow(canvass);
```
